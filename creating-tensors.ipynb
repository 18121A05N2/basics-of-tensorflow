```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import tensorflow as tf
```

1. creating the tensor using data we have ( tf.constant )

```python
print(tf.constant([1,2,3,4.0],shape=[2,2]))
```

2. creating tensor using the zeros and ones is same as above ( tf.ones, tf.zeros )

3. fill in tensor fill does not contain shape argument or other simply use the shape as the first argument.

```python
print(tf.fill([2,2,3],2))
```

4. linspace in tensor default is float because of division

```python
print(tf.linspace(5,9,5))
print(tf.linspace(5,9,6))
```

5. Random tensor generation normal, uniform.

**normal** use standard deviation and mean as parameter and randomly generates the values

image.png

**uniform** will be strictly bound to the minvalue and maxvalue

```python
print(tf.random.normal([2,3,4],mean=2,stddev=1))
print(tf.random.uniform([2,3,4],minval=1,maxval=3))
```

6. shuffle the tensor. ( It shuffle along the first dimension )

```python

tensor = tf.random.normal([2,3,4],2,1)
print("before tensor is shuffled")
print(tensor)
print("after tensor is shuffled")
print(tf.random.shuffle(tensor))
```
